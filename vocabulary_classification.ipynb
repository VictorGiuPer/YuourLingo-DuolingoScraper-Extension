```python
import pandas as pd
```

```python
spanish = pd.read_json("C:\\Users\\victo\\VSCode Folder\\Duolingo\\DuolingoSupport\\Data\\spanish.json")
french = pd.read_json("C:\\Users\\victo\\VSCode Folder\\Duolingo\\DuolingoSupport\\Data\\french.json")

# Rename the columns correctly
spanish.rename(columns={"Spanish": "Learning"}, inplace=True)
french.rename(columns={"French": "Learning"}, inplace=True)

print(spanish.head())
print(french.head())
```

```python
def find_partial_duplicates(df1, df2, column='Translation'):
    """
    Compares two DataFrames for partial or exact matches within comma-separated phrases
    in the specified column and includes the Learning Language words for comparison.
    """
    matches = []

    for idx1, val1 in df1[column].dropna().items():
        phrases1 = [phrase.strip().lower() for phrase in val1.split(",") if phrase.strip()]
        # Fetch the learning language word safely
        learning_word1 = df1.loc[idx1, "Learning"] if "Learning" in df1.columns else ""

        for idx2, val2 in df2[column].dropna().items():
            phrases2 = [phrase.strip().lower() for phrase in val2.split(",") if phrase.strip()]
            learning_word2 = df2.loc[idx2, "Learning"] if "Learning" in df2.columns else ""

            for p1 in phrases1:
                for p2 in phrases2:
                    if p1 in p2 or p2 in p1:
                        matches.append({
                            'df1_learning_word': learning_word1,
                            'df1_full_translation': val1,
                            'df2_learning_word': learning_word2,
                            'df2_full_translation': val2
                        })

    return pd.DataFrame(matches)

duplicates_df = find_partial_duplicates(french, spanish, column="Translation")
```
